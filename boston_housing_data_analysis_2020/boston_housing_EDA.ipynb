# Distribution Plot

```python
rows = 2
cols = 7

fig, ax = plt.subplots(nrows=rows, ncols=cols, figsize=(20,10))

col_names = data.columns
index = 0

for i in range(rows):
    for j in range(cols):
        sns.distplot(data[col_names[index]],ax=ax[i][j])
        index = index + 1
plt.tight_layout()
```

Analysing Price and features correlation

```python
rows = 2
cols = 7

fig, ax = plt.subplots(rows, cols, figsize=(20,10))

col_names = data.columns
index = 0
for i in range(rows):
    for j in range(cols):
        sns.scatterplot(x='Price', y=col_names[index], data=data, ax=ax[i][j], color='darkgreen')
        index += 1
plt.tight_layout()
plt.show()
```
